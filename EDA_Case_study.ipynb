 # DATA CLEANING

### MISSING VALUES

```python
# Checking missing values

loan_df.isnull().sum()
```

```python
# Calculate percentage of missing values for each column

missing_percentages = loan_df.isnull().sum() / len(loan_df) * 100
print(missing_percentages)
```

```python
# Define the threshold (30%)
threshold = 30
```

```python
# Filter columns based on the threshold

filtered_columns = missing_percentages[missing_percentages >= threshold].index
print(filtered_columns)
```

So these are the columns having more than 30% of null values. We will drop these columns from data frame.

```python
# Drop the columns from the original DataFrame

loan_df_filtered = loan_df.drop(columns=filtered_columns)
```

```python
# Now check the shape again.

loan_df = loan_df_filtered
loan_df.shape
```

```python
# Checking missing values again

loan_df.isnull().sum()
```

```python
# Calculate percentage of missing values for each column again

missing_percentages = loan_df.isnull().sum() / len(loan_df) * 100
print(missing_percentages)
```

Some columns like annual_inc, dti, revol_util, pub_rec_bankruptcies, etc., are crucial for assessing a borrower's financial health and creditworthiness. So we can impute the Categorical Columns (title, last_credit_pull_d, last_pymnt_d) with MODE and Numerical Columns (collections_12_mths_ex_med, chargeoff_within_12_mths, tax_liens) with MEDIAN

```python
# Defining the columns with low missing percentages:
columns_to_impute = ['title', 'last_credit_pull_d', 'last_pymnt_d', 'collections_12_mths_ex_med', 'chargeoff_within_12_mths', 'tax_liens']
```

```python
# Imputing categorical columns with mode

for col in ['title', 'last_credit_pull_d', 'last_pymnt_d']:
    loan_df[col].fillna(loan_df[col].mode()[0], inplace=True)
```

```python
# Impute numerical columns with median

for col in ['collections_12_mths_ex_med', 'chargeoff_within_12_mths', 'tax_liens']:
    loan_df[col].fillna(loan_df[col].median(), inplace=True)
```

```python
# Checking missing values after imputation

loan_df.isnull().sum()
```

So we still had few missing values.

Since (emp_title) and (emp_length) are related to employment information, they are categorical and ordinal respectively and (revol_util) and (pub_rec_bankruptcies), which are measures of financial behavior.

We typically imputing missing values of (emp_title) with a generic category like "Unknown" or "Not Provided" and
imputing missing values of (emp_length) with the mode (most common employment length) or a separate category like "Not Provided" and
imputing missing values of (revol_util) with the median to maintain the distribution of utilization rates and imputing missing values of (pub_rec_bankruptcies)
with the mode as it's a discrete variable indicating a count.

```python
# Imputing 'emp_title' with 'Unknown'

loan_df['emp_title'].fillna('Unknown', inplace=True)
```

```python
# Imputing 'emp_length' with a separate category 'Not Provided'

loan_df['emp_length'].fillna('Not Provided', inplace=True)
```

```python
# Imputing 'revol_util' with median
loan_df['revol_util'].fillna(loan_df['revol_util'].median(), inplace=True)
```

```python
# Imputing 'pub_rec_bankruptcies' with mode
loan_df['pub_rec_bankruptcies'].fillna(loan_df['pub_rec_bankruptcies'].mode()[0], inplace=True)
```

```python
# Checking the overall missing values in the DataFrame again after imputing

print("\nTotal missing values in the DataFrame:")
loan_df.isnull().sum()
```

### BIVARIATE ANALYSIS

Bivariate analysis involves examining the relationship between two variables. Given our objective of identifying patterns indicating loan default, we should focus on the relationship between our independent variables (e.g., loan amount, term, interest rate, grade, etc.) and the dependent variable (loan_status)

--> Box plots to compare loan amounts for different loan statuses.

#### Loan Amount vs. Loan Status



```python
# Ensure loan_status is categorized appropriately
loan_df['loan_status'] = loan_df['loan_status'].astype('category')

# Box plot for loan amount vs. Loan status
plt.figure(figsize=(10, 6))
sns.boxplot(x='loan_status', y='loan_amnt', data=loan_df)
plt.title('Loan Amount vs. Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Loan Amount')
plt.xticks(rotation=45)
plt.show()

# Statistical summary for loan amount by loan status
loan_amount_stats = loan_df.groupby('loan_status')['loan_amnt'].describe()
print("\nSummary statistics for Loan Amount vs Loan Status:")
print(loan_amount_stats)

```

#### Insights from Above ####

--> Loans that are currently being paid (Current status) have higher average and median loan amounts compared to both charged-off and fully-paid loans. This suggests that larger loans might be more actively managed or have stricter repayment oversight.

--> Charged-off loans tend to have higher loan amounts than fully-paid loans, indicating that higher loan amounts might pose a higher risk of default.



#### Interest Rate vs. Loan Status

```python
# Box plot for interest rates vs. Loan status
plt.figure(figsize=(10, 6))
sns.boxplot(x='loan_status', y='int_rate', data=loan_df)
plt.title('Interest Rates vs. Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Statistical summary for interest rates by loan status
interest_rate_stats = loan_df.groupby('loan_status')['int_rate'].describe()
print("\nSummary statistics for Interest Rate vs Loan Status:")
print(interest_rate_stats)
```

#### Insights from Above ####

--> Loans that are currently being paid (Current status) exhibit the highest average and median interest rates among all loan statuses.

--> Charged-off loans have higher interest rates on average compared to fully-paid loans. This implies that higher interest rates could potentially correlate with increased default risk.

#### DTI Ratio vs Loan Status

```python
# Ensure loan_status is categorized appropriately
loan_df['loan_status'] = loan_df['loan_status'].astype('category')

# Box plot for DTI ratio vs. Loan status
plt.figure(figsize=(10, 6))
sns.boxplot(x='loan_status', y='dti', data=loan_df)
plt.title('DTI Ratio vs. Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('DTI Ratio')
plt.xticks(rotation=45)
plt.show()

# Statistical summary for DTI ratio by loan status
dti_stats = loan_df.groupby('loan_status')['dti'].describe()
print("\nSummary statistics for DTI Ratio vs Loan Status:")
print(dti_stats)


```

-->Charged-off loans have higher DTI ratios than fully-paid loans, suggesting that borrowers with higher DTI ratios may face greater challenges in loan repayment.

#### Stacked Bar Chart: Grade vs. Loan Status

Used a stacked bar chart to compare the distribution of loan grades across different loan statuses. This will help to see if certain grades are more likely to result in default or full payment.

```python
# Crosstab for Grade vs. Loan Status
grade_status_crosstab = pd.crosstab(loan_df['grade'], loan_df['loan_status'], normalize='index')

# Stacked bar chart for Grade vs. Loan Status
grade_status_crosstab.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Grade vs. Loan Status')
plt.xlabel('Grade')
plt.ylabel('Proportion of Loan Status')
plt.legend(title='Loan Status')
plt.show()


# Statistical summary for Grade by loan status
grade_stats = loan_df.groupby('grade')['loan_status'].value_counts(normalize=True).unstack()
print("\nSummary statistics for Grade vs Loan Status:")
print(grade_stats)
```

#### Insights from Above ####

--> Loans with higher grades (A and B) have a significantly higher proportion of fully-paid loans compared to lower grades (F and G). This indicates that loans with better grades are associated with lower default rates.

--> Lower loan grades (F and G) have a higher proportion of charged-off loans, highlighting an increased risk of default with lower-quality loans.

# CONCLUSION #

### To minimize the risk of loan default, lenders should consider: ###

##### Risk-Based Pricing : Adjusting interest rates according to borrower risk profiles, ensuring that higher-risk borrowers compensate for their potential default risk.

##### Stricter DTI Limits: Implementing tighter debt-to-income ratio thresholds to mitigate default risks associated with higher debt burdens.

##### Enhanced Monitoring: Maintaining vigilance over loans with larger amounts and lower grades, as these factors correlate with higher default rates.

##### Credit Quality: Emphasizing the importance of credit grades in loan approval processes to reduce the likelihood of defaults.
